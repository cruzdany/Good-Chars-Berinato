## Chapter 5.

### 1. Mention the basic structure of every chart. 
Title, subtitle, visual field (visuals, axes, labels, captions, legend), and source line. 
 

### 2. What is the consisten weight needed for each element in charts:

* Title: 12%
* Subtitle: 8%
* Field: 75%
* Source line: 5% 


### 3. Mention the goal of a good chart design

To make them more effective and easier to understand; self-explanatory.


### 4. What things should make charts look elegant or beautiful

Show only what is needed.
Avoid belt and suspenders design.
Minimize the number of colors you use.
Limite eye travel.


### 5. What things should be useful to make sense or understood a chart instantly?

* Remove extraneous elements.
* Make all the elements support the visual.
* Remove ambiguity.
* Use conventions and metaphors.


### 6. Mention the path to simplicity
Removing unnecessary things from the chart, leaving only what is valuable to communicating your message.

## Chapter 7.

### 1. What is the double Y-axis?
It is a chart that includes two vertical scales for different datasets in the visual field.

### 2. Why the double Y-axis might be effective?
Since it compels the viewer to make comparisons between datasets.

### 3. Why the double Y-axis might be deceptive?
Since relationships between different values are artificial.

### 4. What is a map?
It is a representation that uses geographical boundaries to encode values related to a location

### 5. Why the map might be deceptive?
Since the size of a region does not neccesarily reflect the data encoded within it

### 6. What is the truncated Y-axis?
It is a chart that removes valid value range from the y-axis.
